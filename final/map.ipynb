```python
import datetime
from tqdm import tqdm
```

```python
df['YDate'] = None
```

```python
dt1 = datetime.datetime(2019,1,1,0,)#Y,M,D,H
for idx, data in tqdm(df.iterrows()):
    dt2 = datetime.datetime(2019,data['Month'],data['Mdate'],data['Time'])
    dur = dt2 - dt1
    df.loc[idx,'YDate'] = dur.total_seconds()
```

```python
df['YDate'] = df['YDate']/3600
```
